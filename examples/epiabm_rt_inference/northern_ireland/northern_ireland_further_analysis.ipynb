This notebook is for making further plots from the inference and optimisation performed in the `northern_ireland_rt_inference.ipynb` notebook within this directory. Here we will compare the inference of varying levels of parameters.

We now wish to capture all of the above information into two separate figures. ### EpiABM figure

```python
import matplotlib.pyplot as plt
from matplotlib import image
import numpy as np
import pandas as pd
```

```python
radius = "r_1"
seir_path = f"NI_outputs/{radius}/SEIR.png"
spatial_NI_path = f"NI_outputs/{radius}/output_NI_grid.png"
Rt_path = f"NI_outputs/{radius}/secondary_infections_mesh.png"
paths_epiabm = [seir_path, spatial_NI_path, Rt_path]
```

### Sampling results all on one plot

```python
times = np.arange(0, 91, 1)
Rt_3on_1off = pd.read_csv("inference_outputs/r_1/bgk_AR1_prev_react_3_1/Rt_info.csv", index_col=0)
Rt_2on_2off = pd.read_csv("inference_outputs/r_1/bgk_AR1_prev_react_2_2/Rt_info.csv", index_col=0)
Rt_1on_3off = pd.read_csv("inference_outputs/r_1/bgk_AR1_prev_react_1_3/Rt_info.csv", index_col=0)
true_R_t = Rt_3on_1off["true"]
```

```python
plt.figure(figsize=(8, 2.5))
plt.plot(times, true_R_t, 'k', 
         label="True $R_t$")
plt.plot(times, Rt_3on_1off["mean"], color="blue")
plt.fill_between(times, Rt_3on_1off["lower"], Rt_3on_1off["upper"], alpha=0.5,
                 label="3 weeks on, 1 week off", color="blue")
plt.plot(times, Rt_2on_2off["mean"], color="green")
plt.fill_between(times, Rt_2on_2off["lower"], Rt_2on_2off["upper"], alpha=0.5,
                 label="2 weeks on, 2 weeks off", color="green")
plt.plot(times, Rt_1on_3off["mean"], color="goldenrod")
plt.fill_between(times, Rt_1on_3off["lower"], Rt_1on_3off["upper"], alpha=0.5,
                 label="1 week on, 3 weeks off", color="goldenrod")
plt.xlabel("Time [days]")
plt.ylabel("$R_t$")
# plt.title("Inferred $R_t$ for Northern Ireland simulation: Case 2")
plt.ylim(0, 16)
plt.legend()
# plt.colorbar(cf)
# plt.show()
plt.savefig("inference_outputs/r_1/sampling_periods.png", bbox_inches="tight")
```
